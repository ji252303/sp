```python
# fastapi_app.py

from fastapi import FastAPI, HTTPException
import openai
import requests
import deepl

app = FastAPI()

# GPT 및 Deepl API 키 설정
openai.api_key = ""
DEEPL_API_KEY = ""
DEEPL_API_URL = 'https://api-free.deepl.com/v2/translate'

def translate_text(text, target_lang):
    data = {
        'auth_key': DEEPL_API_KEY,
        'text': text,
        'target_lang': target_lang
    }
    response = requests.post(DEEPL_API_URL, data=data)
    if response.status_code != 200:
        raise HTTPException(status_code=response.status_code, detail="Translation API error")
    return response.json()['translations'][0]['text']

def get_gpt_response(prompt):
    response = openai.ChatCompletion.create(
        model="ft:gpt-3.5-turbo-0125:personal::9qyCrJtR",
        messages=[
            {"role": "user", "content": prompt}
        ]
    )
    return response.choices[0].message.content

@app.post("/query/")
async def handle_user_query(user_query: str):
    try:
        # 사용자 질문을 한국어로 번역
        translator = deepl.Translator(DEEPL_API_KEY)
        result = translator.translate_text(user_query, target_lang="KO")
        translated_text = result.text

        # GPT에게 사용자 질문 전달
        gpt_response_korean = get_gpt_response(translated_text)

        # GPT의 한국어 답변을 중국어로 번역
        gpt_response_chinese = translate_text(gpt_response_korean, target_lang="ZH")

        # 한국어 및 중국어 답변 반환
        return {
            "korean_response": gpt_response_korean,
            "chinese_response": gpt_response_chinese
        }
    except Exception as e:
        raise HTTPException(status_code=500, detail=str(e))
```
